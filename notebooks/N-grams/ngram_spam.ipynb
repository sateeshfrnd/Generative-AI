# N-gram

An N-gram is a contiguous sequence of N words or characters from a given text. It helps capture local word dependencies and partial context in text data.

```python
import pandas as pd
messages=pd.read_csv('spam.csv', delimiter=',',usecols=[0, 1])
```

```python
messages.head()
```

## Data Cleaning And Preprocessing

```python
import re
import nltk
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
corpus=[]
for i in range(0,len(messages)):
    review=re.sub('[^a-zA-z]',' ',messages['message'][i])
    review=review.lower()
    review=review.split()
    review=[ps.stem(word) for word in review if not word in stopwords.words('english')]
    review=' '.join(review)
    corpus.append(review)
```

```python
corpus[:10]
```

## Create Bag Of Words

```python
from sklearn.feature_extraction.text import CountVectorizer

# Initialize CountVectorizer for unigrams (1-grams)
cv = CountVectorizer(max_features=100,binary=True, ngram_range=(1,1))
```

```python
# Fit and transform the sentences
ngram_matrix=cv.fit_transform(corpus).toarray()
```

```python
cv.vocabulary_
```

```python
# Initialize CountVectorizer for bigrams 
cv2 = CountVectorizer(max_features=100,binary=True, ngram_range=(2,2))
# Fit and transform the sentences
ngram2_matrix=cv2.fit_transform(corpus).toarray()
```

```python
cv2.vocabulary_
```

```python
# Initialize CountVectorizer for trigrams 
cv3 = CountVectorizer(max_features=100,binary=True, ngram_range=(3,3))
# Fit and transform the sentences
ngram3_matrix=cv3.fit_transform(corpus).toarray()
```

```python
cv3.vocabulary_
```

```python
import numpy as np
np.set_printoptions(edgeitems=30, linewidth=100000, 
    formatter=dict(float=lambda x: "%.3g" % ngram3_matrix))
ngram3_matrix
```
